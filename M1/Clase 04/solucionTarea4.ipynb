A partir del CSV hospitales2.csv, generar un archivo CSV de salida, que contenga los siguientes campos en este orden:
latitude
longitude
name
label
Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder con la dirección del hospital, y el campo label con el nombre del hospital.

```python
import pandas as pd
```

```python
datosHospital = pd.read_csv('hospitales2.csv')
datosHospital.head(3)
```

```python
def encontrarCordenadas(point):
    point = point.split()[1:]

    longitud = float(point[0].replace('(', ''))
    latitud = float(point[1].replace(')', ''))

    return longitud, latitud
```

Aplicar longitud y latitud al dataframe 

```python
#creando culimnas latitud y longitud 
datosHospital['longitud'] = datosHospital['WKT'].apply(lambda x: encontrarCordenadas(x)[0])
datosHospital['latitud'] = datosHospital['WKT'].apply(lambda x: encontrarCordenadas(x)[1])

datosHospital.head(1)
```

```python
#crear nuevo Dataframe
df_salida = datosHospital[[ 'latitud', 'longitud', 'DOM_NORMA', 'NOMBRE']]
df_salida = df_salida.rename(columns={'DOM_NORMA':'name', 'NOMBRE':'label'})

#quitar mallusculas a columnas
df_salida['name'] = df_salida['name'].str.title()
df_salida['label'] = df_salida['label'].str.title()
df_salida.head(3)
```

```python

df_salida.to_csv('SolucionTarea4.csv', index=False)
```
